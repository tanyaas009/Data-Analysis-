```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("setup complete")
```

```python
#path of file to read

candy_filepath="candy.csv"
#read the data into a variable
candy_data= pd.read_csv(candy_filepath, index_col='id')
print("setup complete")
```

```python
candy_data.head(30)
```

```python
#  Which candy was more popular with survey respondents:
# '3 Musketeers' or 'Almond Joy'?  
musk=candy_data[candy_data.competitorname=='3 Musketeers']['winpercent'].item()
alm=candy_data[candy_data.competitorname=='Almond Joy']['winpercent'].item()
more_popular = '3 Musketeers' if musk>alm else 'Almond Joy'

#  Which candy has higher sugar content: 'Air Heads'
# or 'Baby Ruth'? 
airsug=candy_data[candy_data.competitorname=='Air Heads']['sugarpercent'].item()
babysug=candy_data[candy_data.competitorname=='Baby Ruth']['sugarpercent'].item()
more_sugar = 'Air Heads' if airsug>babysug else 'Baby Ruth'
```

```python
# Scatter plot showing the relationship between 'sugarpercent' and 'winpercent'
sns.scatterplot(x=candy_data.sugarpercent, y=candy_data.winpercent)

#The scatter plot does not show a strong correlation between the two variables. Since there is no clear relationship between the two variables, this tells us that sugar content does not play a strong role in candy popularity.
```

```python
# Scatter plot w/ regression line showing the relationship between 'sugarpercent' and 'winpercent'
sns.regplot(x=candy_data.sugarpercent, y=candy_data.winpercent)
#Since the regression line has a slightly positive slope, this tells us that there is a slightly positive correlation between `'winpercent'` and `'sugarpercent'`. Thus, people have a slight preference for candies containing relatively more sugar.
```

```python
# Color-coded scatter plot w/ regression lines
sns.lmplot(x='sugarpercent', y='winpercent', hue='chocolate', data=candy_data)
#with the regression line for chocolate candies. Since this line has a slightly positive slope, we can say that more expensive chocolatecandies tend to be more popular (than relatively cheaper chocolate candies).  Likewise, since the regression line for candies without chocolate has a negative slope, we can say that if candies don't contain chocolate, they tend to be more popular when they are cheaper. 
```
